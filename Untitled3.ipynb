```python
import pandas as pd
import requests
import re
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

# =========================
# LOAD CSV
# =========================

csv_path = "/content/Chatbot_500_Rows_India_Weather_Map_GK_Cricket_Calculator.csv"

df = pd.read_csv(csv_path)
df = df.dropna()

print("Dataset loaded:", df.shape)

# =========================
# EMBEDDING MODEL
# =========================

model = SentenceTransformer("all-MiniLM-L6-v2")

questions = df["input"].astype(str).str.lower().str.strip().tolist()

question_embeddings = model.encode(
    questions,
    convert_to_numpy=True,
    normalize_embeddings=True
)

dimension = question_embeddings.shape[1]

index = faiss.IndexFlatIP(dimension)
index.add(question_embeddings)

print("Vector database ready!")

# =========================
# WEATHER API
# =========================

def get_weather(city):
    api_key = "0a44b1e5f46e474d85021542f76525f3"

    url = f"https://api.openweathermap.org/data/2.5/weather?q={city},IN&appid={api_key}&units=metric"

    try:
        data = requests.get(url).json()

        if data.get("cod") != 200:
            return f"Weather data for {city.title()} not found."

        temp = data["main"]["temp"]
        feels_like = data["main"]["feels_like"]
        humidity = data["main"]["humidity"]
        desc = data["weather"][0]["description"]

        return f"Weather in {city.title()}: {temp}°C, feels like {feels_like}°C, {desc}, humidity {humidity}%."

    except:
        return "Weather API error."

# =========================
# MAP
# =========================

def get_map_link(place):
    place_url = place.replace(" ", "+")
    return f"Map of {place.title()}: https://www.google.com/maps/search/{place_url}"

# =========================
# CALCULATOR
# =========================

def calculate_expression(query):
    try:
        expression = query.lower()

        replacements = {
            "calculate": "",
            "what is": "",
            "solve": "",
            "plus": "+",
            "minus": "-",
            "into": "*",
            "multiply": "*",
            "multiplied by": "*",
            "divide by": "/",
            "divided by": "/",
            "power": "**"
        }

        for old, new in replacements.items():
            expression = expression.replace(old, new)

        expression = expression.strip()

        if not re.match(r"^[0-9+\-*/().\s]+$", expression):
            return None

        answer = eval(expression)
        return f"The answer is {answer}"

    except:
        return None

# =========================
# INTENT FUNCTIONS
# =========================

def is_weather_query(query):
    return "weather" in query or "temperature" in query

def extract_weather_city(query):
    query = query.lower()
    for word in ["weather in", "temperature in", "weather of", "temperature of", "weather", "temperature"]:
        query = query.replace(word, "")
    return query.strip()

def is_map_query(query):
    return query.startswith("map of") or query.startswith("location of") or query.startswith("where is")

def extract_place(query):
    query = query.lower()
    for word in ["map of", "location of", "where is"]:
        query = query.replace(word, "")
    return query.strip()

def is_calculator_query(query):
    return any(op in query for op in ["calculate", "solve", "plus", "minus", "+", "-", "*", "/"])

# =========================
# CHATBOT
# =========================

print("\nChatbot ready! Type exit to stop.")

while True:
    query = input("\nYou: ").lower().strip()

    if query in ["exit", "quit", "bye"]:
        print("Bot: Bye!")
        break

    if is_weather_query(query):
        city = extract_weather_city(query)
        print("Bot:", get_weather(city))
        continue

    if is_map_query(query):
        place = extract_place(query)
        print("Bot:", get_map_link(place))
        continue

    if is_calculator_query(query):
        ans = calculate_expression(query)
        if ans:
            print("Bot:", ans)
            continue

    query_embedding = model.encode(
        [query],
        convert_to_numpy=True,
        normalize_embeddings=True
    )

    scores, indexes = index.search(query_embedding, k=1)

    best_score = scores[0][0]
    best_index = indexes[0][0]

    if best_score > 0.45:
        print("Bot:", df.iloc[best_index]["response"])
        print("Category:", df.iloc[best_index]["category"])
        print("Matched Question:", df.iloc[best_index]["input"])
        print("Score:", round(float(best_score), 2))
    else:
        print("Bot: Sorry, I don't know this. Add this question to dataset.")
```
